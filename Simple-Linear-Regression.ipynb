```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm

# Load the dataset
df = pd.read_csv('Study_Hours.csv')

# Display the first few rows of the dataset
print("Dataset Preview:")
print(df.head())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Define features and target
X = df[['Hours of Study']]  # Feature
y = df['Exam Score']        # Target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create and train the model using scikit-learn
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Display scikit-learn results
print("\nscikit-learn Model Coefficients:")
print(f"Intercept (b0): {model.intercept_}")
print(f"Coefficient (b1): {model.coef_[0]}")
print("\nscikit-learn Model Performance:")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
print(f"R-squared: {r2_score(y_test, y_pred)}")

# Add a constant (intercept) to the model for statsmodels
X_train_sm = sm.add_constant(X_train)

# Create and fit the model using statsmodels
sm_model = sm.OLS(y_train, X_train_sm).fit()

# Display the statsmodels summary
print("\nstatsmodels OLS Regression Results:")
print(sm_model.summary())

# Calculate residuals
residuals = y_test - y_pred

# Generate Q-Q plot
sm.qqplot(residuals, line='45')
plt.title('Q-Q Plot of Residuals')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()

# Visualize the regression results
plt.scatter(X_test, y_test, color='blue', label='Actual Scores')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted Scores')
plt.title('Hours of Study vs. Exam Score')
plt.xlabel('Hours of Study')
plt.ylabel('Exam Score')
plt.legend()
plt.show()

```

Solution if the data is not normally distributed

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import boxcox
import statsmodels.api as sm

# Load the dataset
df = pd.read_csv('Study_Hours.csv')

# Display the first few rows of the dataset
print("Dataset Preview:")
print(df.head())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Define features and target
X = df[['Hours of Study']]
y = df['Exam Score']

# Step 1: Check for Outliers using Boxplot
plt.boxplot(y)
plt.title('Boxplot of Exam Score')
plt.ylabel('Exam Score')
plt.show()

# Optionally, you can remove outliers based on the IQR method
Q1 = y.quantile(0.25)
Q3 = y.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(y >= lower_bound) & (y <= upper_bound)]

# Step 2: Transform the Data
# Applying Box-Cox transformation to y (only for positive data)
y_transformed, _ = boxcox(y[y > 0])
plt.hist(y_transformed, bins=10, color='blue', alpha=0.7)
plt.title('Histogram of Transformed Exam Score (Box-Cox)')
plt.show()

# Step 3: Fit a Non-Linear Model (Polynomial Regression)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
polynomial_features = PolynomialFeatures(degree=2)
X_poly_train = polynomial_features.fit_transform(X_train)
X_poly_test = polynomial_features.transform(X_test)

poly_model = LinearRegression()
poly_model.fit(X_poly_train, y_train)

# Predictions
poly_y_pred = poly_model.predict(X_poly_test)

# Performance metrics for polynomial regression
print("Polynomial Regression:")
print(f"Mean Squared Error: {mean_squared_error(y_test, poly_y_pred)}")
print(f"R-squared: {r2_score(y_test, poly_y_pred)}")

# Step 4: Robust Regression (Quantile Regression)
from statsmodels.regression.quantile_regression import QuantReg

quantile_model = QuantReg(y_train, sm.add_constant(X_train)).fit(q=0.5)
print("\nQuantile Regression Summary:")
print(quantile_model.summary())

# Predictions using Quantile Regression
quantile_y_pred = quantile_model.predict(sm.add_constant(X_test))

# Step 5: Visualizations
plt.scatter(X_test, y_test, color='blue', label='Actual Scores')
plt.scatter(X_test, quantile_y_pred, color='green', label='Quantile Predicted Scores')
plt.plot(X_test, poly_y_pred, color='red', linewidth=2, label='Polynomial Regression')
plt.title('Regression Models Comparison')
plt.xlabel('Hours of Study')
plt.ylabel('Exam Score')
plt.legend()
plt.show()
```
